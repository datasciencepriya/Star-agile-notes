```python
cols = ['Vehicle Class', 'Fuel Type', 'Transmission', 'Cylinders', 'Engine Size(L)', 'Fuel Consumption Comb (L/100 km)']
X = data[cols]                      # Independent Feature Set
Y = data[['CO2 Emissions(g/km)']]   # Depedent Feature
```

```python
X.head(10)
```

```python
Y.head(10)
```

```python
print('Independent Feature Set Shape : ', X.shape)
print('Dependent Feature Shape       : ', Y.shape)
```

```python
# Splitting The Data into training and test sets
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X,Y, test_size = 0.2, random_state = 2)
```

```python
print('Training Data Shape   : ', x_train.shape)
print('Training Labels Shape : ', y_train.shape)
print('Testing Data Shape    : ', x_test.shape)
print('Testing Labels Shape  : ', y_test.shape)
```

```python
# Encoding (Dealing with Categorical Data)
cols = ['Vehicle Class', 'Fuel Type', 'Transmission']
from sklearn.preprocessing import OrdinalEncoder
encoder = OrdinalEncoder()

# Fit on train set
encoder.fit(x_train[cols])

# Tranforming the training and testing sets
x_train[cols] = encoder.transform(x_train[cols])
x_test[cols]  = encoder.transform(x_test[cols])
```

```python
x_train.head(10)
```

```python
# Scaling
cols = ['Engine Size(L)', 'Fuel Consumption Comb (L/100 km)']
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

# Fit on training set
scaler.fit(x_train[cols])

# Tranform training and testing sets
x_train[cols] = scaler.transform(x_train[cols])
x_test[cols]  = scaler.transform(x_test[cols])
```

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor()
```

```python
search_params = {
    'n_estimators' : [10,20,30,40,50],
    'max_depth'    : [3,4,5,6,8,10],
    'max_features' : [3,4,5,6]
}
```

```python
from sklearn.model_selection import GridSearchCV
grid_search = GridSearchCV(estimator = regressor,
                           param_grid = search_params,
                           scoring = ['r2'],
                           cv = 5,
                           refit = 'r2',
                           verbose = 1)
grid_search.fit(x_train, y_train)
```

```python
grid_search.best_params_
```

```python
grid_search.best_score_
```

```python
best_model = grid_search.best_estimator_
best_model
```

```python
grid_search.cv_results_
```

```python
df = pd.DataFrame(grid_search.cv_results_).sort_values('mean_test_r2', ascending = False)
df.head()
```

```python
cols = ['param_max_depth','param_max_features','param_n_estimators','params','mean_test_r2','rank_test_r2']
df[cols].to_csv('GridSearchResults.csv', index = False)
```

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor()
```

```python
search_params = {
    'n_estimators' : [10,20,30,40,50],
    'max_depth'    : [3,4,5,6,8,10],
    'max_features' : [3,4,5,6]
}
```

```python
from sklearn.model_selection import RandomizedSearchCV
randomized_search = RandomizedSearchCV(estimator = regressor,
                                       param_distributions = search_params,
                                       scoring = ['r2'],
                                       cv = 5,
                                       refit = 'r2',
                                       n_iter = 20,
                                       verbose = 3)
randomized_search.fit(x_train, y_train)
```

```python
randomized_search.best_params_
```

```python
randomized_search.best_score_
```

```python
best_model = randomized_search.best_estimator_
best_model
```

```python
df = pd.DataFrame(randomized_search.cv_results_).sort_values('mean_test_r2', ascending = False)
df.head()
```

```python
cols = ['param_max_depth','param_max_features','param_n_estimators','params','mean_test_r2','rank_test_r2']
df[cols].to_csv('RandomizedSearchResults.csv', index = False)
```
