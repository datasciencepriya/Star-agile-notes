```python
cols = ['BP', 'Cholesterol', 'Sex', 'Age', 'Na_to_K']
X = data[cols]
Y = data[['Drug']]

print('Independent Feature Set Shape : ', X.shape)
print('Dependent Feature Shape       : ', Y.shape)
```

```python
X.head()
```

```python
Y.head()
```

```python
# Encoding
X.Sex = X.Sex.map({'M': 0, 'F': 1})
X.Cholesterol = X.Cholesterol.map({'NORMAL': 0, 'HIGH': 1})
X.BP = X.BP.map({'LOW': 0, 'NORMAL': 1, 'HIGH': 2})

X.head(10)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2, random_state = 5)
```

```python
print('Training Data Shape   : ', x_train.shape)
print('Training Labels Shape : ', y_train.shape)
print('Testing Data Shape    : ', x_test.shape)
print('Testing Labels Shape  : ', y_test.shape)
```

---
---

```python
# Training The Model
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p = 2)
classifier.fit(x_train, y_train)
```

```python
# Evaluate the model
pred_train = classifier.predict(x_train)
pred_test  = classifier.predict(x_test)

from sklearn.metrics import accuracy_score
print('Training Set Accuracy : ', accuracy_score(y_train, pred_train))
print('Testing Set Accuracy  : ', accuracy_score(y_test, pred_test))
```

## Effect of Value of K(n_neighbors)
- When the value of k is very low then we have a complex and inappropriate model which leads to overfitting
- When the value of k is very high then we have an over generalized model which leads to underfitting

```python
# Very low value of k
classifier = KNeighborsClassifier(n_neighbors = 1, metric = 'minkowski', p = 2)
classifier.fit(x_train, y_train)

pred_train = classifier.predict(x_train)
pred_test = classifier.predict(x_test)

print('Training Set Accuracy : ', accuracy_score(y_train, pred_train))
print('Testing Set Accuracy  : ', accuracy_score(y_test, pred_test))
```

```python
# very high value of k
classifier = KNeighborsClassifier(n_neighbors = 15, metric = 'minkowski', p = 2)
classifier.fit(x_train, y_train)

pred_train = classifier.predict(x_train)
pred_test = classifier.predict(x_test)

print('Training Set Accuracy : ', accuracy_score(y_train, pred_train))
print('Testing Set Accuracy  : ', accuracy_score(y_test, pred_test))
```

## Selecting the best value of k(n_neighbors)

```python
acc_train = []
acc_test  = []

for k in range(1,21):
    classifier = KNeighborsClassifier(n_neighbors = k, metric = 'minkowski', p = 2)
    classifier.fit(x_train, y_train)
    
    pred_train = classifier.predict(x_train)
    pred_test  = classifier.predict(x_test)
    
    acc_train.append(accuracy_score(y_train, pred_train))
    acc_test.append(accuracy_score(y_test, pred_test))
```

```python
k = range(1,21)

plt.figure(figsize = (16,8), dpi = 300)
plt.plot(k, acc_train, '-o', label = 'Training Accuracy')
plt.plot(k, acc_test, '--o', label = 'Testing Accuracy')

plt.xticks(k)
plt.xlabel('Number of neighbors(k)')
plt.ylabel('Model performace wrt k')
plt.legend()
```

```python
# Best model (model with value of k = 3)
classifier = KNeighborsClassifier(n_neighbors = 3, metric = 'minkowski', p = 2)
classifier.fit(x_train, y_train)

pred_train = classifier.predict(x_train)
pred_test  = classifier.predict(x_test)

print('Training Set Accuracy : ', accuracy_score(y_train, pred_train))
print('Testing Set Accuracy  : ', accuracy_score(y_test, pred_test))
```
