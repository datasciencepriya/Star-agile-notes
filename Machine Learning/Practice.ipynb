```python
# What percentage of patients having high cholesterol are having high BP?
high_chol = len(data[data.Cholesterol == 'HIGH'])
high_chol_high_bp = len(data[(data.Cholesterol == 'HIGH') & (data.BP == 'HIGH')])
print('Patients having high cholesterol : ', high_chol)
print('Patients having high BP out of patients having high cholesterol : ', high_chol_high_bp)
high_chol_high_bp / high_chol * 100
```

```python
high_chol = data[data.Cholesterol == 'HIGH']
high_chol.BP.value_counts(normalize = True)
```

```python
# Which Gender is more likely to have High Cholesterol problem?
plt.figure(figsize = (16,6), dpi = 200)
sns.countplot('Sex', hue = 'Cholesterol', data = data)
```

```python
data.groupby('Sex').Cholesterol.value_counts(normalize = True)
```

```python
# Which Gender is more likely to have high BP problem?
data.groupby('Sex').BP.value_counts(normalize = True)
```

```python
# 4. In What Age Group High BP and High Cholesterol problems are more common?
plt.figure(figsize = (16,6), dpi = 200)
sns.violinplot('BP', 'Age', data = data)
```

```python
temp_data = data[['Age', 'BP', 'Cholesterol']]
temp_data.head()
```

```python
# Binning (Creating Bands)
temp_data.loc[temp_data.Age <= 30, 'Age'] = 0
temp_data.loc[(temp_data.Age > 30) & (temp_data.Age <= 55), 'Age'] = 1
temp_data.loc[temp_data.Age > 55, 'Age'] = 2

temp_data.head(10)
```

```python
plt.figure(figsize = (16,6), dpi = 200)
sns.countplot('Age', hue = 'BP', data = temp_data)
```

```python
# Is high Na_to_K a contributing factor towards High Cholesterol
plt.figure(figsize = (16,6), dpi = 200)
plt.subplot(1,2,1)
sns.barplot('Cholesterol', 'Na_to_K', data = data)
plt.subplot(1,2,2)
sns.boxplot('Cholesterol', 'Na_to_K', data = data)
```
