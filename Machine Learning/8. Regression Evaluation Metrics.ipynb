```python
cols = ['Vehicle Class', 'Fuel Type', 'Transmission', 'Cylinders', 'Engine Size(L)', 'Fuel Consumption Comb (L/100 km)']
X = data[cols]
Y = data[['CO2 Emissions(g/km)']]
```

```python
# Independent feature Set
X.head(10)
```

```python
# Dependent Feature
Y.head(10)
```

```python
print('Independent Feature Set Shape : ', X.shape)
print('Dependent Feature Shape       : ', Y.shape)
```

```python
# Splitting Data into training and testing sets
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2, random_state = 2)
```

```python
print('Training data shape   : ', x_train.shape)
print('Training labels shape : ', y_train.shape)
print('Testing data shape    : ', x_test.shape)
print('Testing labels shape  : ', y_test.shape)
```

```python
# Scaling  (Data Leakage)
cols = ['Engine Size(L)','Fuel Consumption Comb (L/100 km)']
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

# Fitting the object on training set
scaler.fit(x_train[cols])

# Transforming the training and testing sets
x_train[cols] = scaler.transform(x_train[cols])
x_test[cols]  = scaler.transform(x_test[cols])
```

```python
# Encoding
cols = ['Vehicle Class', 'Fuel Type', 'Transmission']
from sklearn.preprocessing import OrdinalEncoder
encoder = OrdinalEncoder()

# Fit on the training set
encoder.fit(x_train[cols])

# Transforming training and test sets
x_train[cols] = encoder.transform(x_train[cols])
x_test[cols] = encoder.transform(x_test[cols])
```

```python
x_train.head()
```

```python
x_test.head()
```

---
# ================================================

### Training the model

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train, y_train)
```

```python
pred_train = regressor.predict(x_train)
pred_test  = regressor.predict(x_test)
```

----
----
## Regression Evaluation Metrics

## Mean Absolute Error
<img align = 'left' src = 'MAE.png' style = 'width:400px;height:100px'/>

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(y_test, pred_test)
```

## Mean Squared Error
<img align = 'left' src = 'MSE.png' style = 'width:400px;height:100px'/>

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test, pred_test)
```

## Root Mean Squared Error
- Square Root of MSE

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test, pred_test, squared = False)
```

```python
328.35060135654004 ** 0.5
```

## R Squared Score (Coefficient Of Determination)
<img align = 'left' src = 'R_square.png' style = 'width:500px;height:100px'/>

```python
from sklearn.metrics import r2_score
r2_score(y_test, pred_test)
```

```python
# The score function available with regressor object also calculates r2_score
regressor.score(x_test, y_test)
```
