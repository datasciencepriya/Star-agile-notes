```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import random
random.seed = 5

from matplotlib import style
style.use('ggplot')

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('adult.csv')
df.head()
```

```python
df.info()
```

```python
plt.figure(figsize = (12,5), dpi = 200)
sns.distplot(df.age, hist = False)
```

```python
print('The number of observations : ', df.age.shape)
```

```python
sample = random.sample(range(0, 48842), 10)
sample
```

```python
df.iloc[sample, 0]
```

```python
n = 100                  # sample size
N = 1000                 # number of samples
sample_means = []
for i in range(N):
    indexes = random.sample(range(48842), n)
    sample  = df.iloc[indexes, 0]
    sample_means.append(sample.mean())
```

```python
plt.figure(figsize = (12,5), dpi = 200)
sns.distplot(sample_means, hist = False)
```

```python
# Mean of sample means is aproximately equal to population mean
population_mean = df.age.mean()
mean_of_samples = np.mean(sample_means)

print('Mean of sample means : ', mean_of_samples)
print('Mean_of_population   : ', population_mean)
```

```python
# Sample standard deviation is approximately equal to (std of population/sqrt of sample size)
std_of_samples = np.std(sample_means)
std_of_population = df.age.std()
calculated_std = std_of_population / np.sqrt(n)

print('Calculated standard deviation of samples : ', calculated_std)
print('Actual Standard deviation of samples     : ', std_of_samples)
```
