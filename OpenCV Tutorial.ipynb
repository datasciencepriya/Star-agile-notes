```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
%matplotlib inline
```

```python
# Reading an image
panda = cv2.imread('panda.jpg')
plt.imshow(panda)
plt.axis('off')
```

```python
koala = cv2.imread('koala.jpg')
plt.imshow(koala)
plt.axis('off')
```

```python
# By default iamges in cv2 are in BGR
# Changing color format
koala_orig = cv2.cvtColor(koala, cv2.COLOR_BGR2RGB)
plt.imshow(koala_orig)
plt.axis('off')
```

```python
# Converting an image to grayscale
panda_gray = cv2.cvtColor(panda, cv2.COLOR_BGR2GRAY)
plt.imshow(panda_gray, 'gray')
plt.axis('off')
```

```python
koala_gray = cv2.cvtColor(koala_orig, cv2.COLOR_RGB2GRAY)
plt.imshow(koala_gray, 'gray')
plt.axis('off')
```

```python
# Extracting edges / pattern
panda2 = cv2.Canny(panda, 150,200)
plt.imshow(panda2, 'gray')
plt.axis('off')
```

```python
# highlighting the edges
panda3 = cv2.dilate(panda2, kernel = (3,3), iterations = 5)
plt.imshow(panda3, 'gray')
plt.axis('off')
```

```python
# Erode (reverse of dilate)
panda4 = cv2.erode(panda3, kernel = (3,3), iterations = 10)
plt.imshow(panda4, 'gray')
plt.axis('off')
```

## Resizing and Cropping

```python
panda = cv2.imread('panda.jpg')
plt.imshow(panda)
plt.axis('off')
```

```python
panda.shape
```

```python
panda2 = cv2.resize(panda, (2000, 1400))
plt.imshow(panda2)
plt.axis('off')
```

```python
panda2.shape
```

```python
# Saving an image
cv2.imwrite('resized_panda.jpg', panda2)
```

```python
# Cropping
panda_cropped = panda[200:600, 400:1000]
plt.imshow(panda_cropped)
plt.axis('off')
```

## Drawing a Bounding Box(Rectangle)

```python
panda = cv2.imread('panda.jpg')
panda = cv2.cvtColor(panda, cv2.COLOR_BGR2RGB)
plt.imshow(panda)
plt.axis('off')
```

```python
cv2.rectangle(panda, (200,100), (1000,700), (255,0,0), 10)
plt.imshow(panda)
plt.axis('off')
```

## Writing text on images

```python
koala = cv2.imread('koala.jpg')
koala = cv2.cvtColor(koala, cv2.COLOR_BGR2RGB)
cv2.putText(koala, 'What are you looking at , let me have my food', (60,600), cv2.FONT_HERSHEY_COMPLEX, 1.2, (255,0,0), 3) 

plt.figure(figsize = (20,14))
plt.imshow(koala)
plt.axis('off')
```
